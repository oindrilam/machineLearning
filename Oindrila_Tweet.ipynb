```python
import langdetect
import langid
import re
def remove_xml(text):
    #return re.sub(r'[-()\"#/@;:<>{}=~|.?,]', ' ', text)
    return re.sub("[^a-zA-Z]",  # Search for all non-letters
                          " ",          # Replace all non-letters with spaces
                          str(text))

def remove_newlines(text):
    return text.replace('\n', ' ') 
    

def remove_manyspaces(text):
    return re.sub(r'\s+', ' ', text)

def clean_text(text):
    text = remove_xml(text)
    text = remove_newlines(text)
    text = remove_manyspaces(text)
    return text
```

```python

from textblob import TextBlob
from langdetect import detect
import csv

languages_langdetect = []

#Loading dataset  in tweets.
tweets    = df1_tidy['text']

for line in tweets:
    
    line=clean_text(line)
    
# the try except blook because there is some tweets contain links
for line in tweets:
    try:
        result = langdetect.detect_langs(line)
        print(result)
        result = str(result[0])[:2]
    except:
        result = 'unknown'
    
    finally:
        languages_langdetect.append(result)
df1_tidy['languages_langdetect'] = languages_langdetect





```
